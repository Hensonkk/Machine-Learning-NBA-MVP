```python
# Load in the modules needed for preprocessing the Data

import pandas as pd
import numpy as np
```

```python
# Read the CSV file for All NBA Season statistics into a Pandas Dataframe

nba_df = pd.read_csv("Datasets/all_nba_seasons.csv")
nba_df.head()
```

```python
# Drop unwanted columns from the Dataframe

nba_df = nba_df.drop(columns = ["FG", "FGA", "3P", "3PA", "3P%", "2P", "2PA", "2P%", "FT", "FTA", "ORB", "DRB", "TOV"])
nba_df.head()
```

```python
# Filter out players by only selecting the players who score more than 5 points a game. 

nba_df2 = nba_df[nba_df["PTS"] > 5]
nba_df2
```

```python
# Drop the duplicate players in each season, this is due to some players being traded and names appearing
# multiple times in one season

nba_df2.drop_duplicates(subset = ['Season', 'Player'], keep = 'first', inplace = True)
```

```python
# Create a new column for MVP and assign non value to it

nba_df2["MVP"] = np.nan
```

```python
# Display dataframe after adjusments are made

nba_df2
```

```python
# Set the column order

columns_order = ["Season", "Player", "Pos", "Age", "G", "MP", "PTS", "AST", "TRB", "STL", "BLK", "FG%", "eFG%", "FT%", "PF", "WS", "WS/48", "MVP"]
```

```python
# Restructure the dataframe to the desired column order

nba_df2 = nba_df2[columns_order]
nba_df2
```

```python
# Export the dataframe into a csv file in our Datasets folder

nba_df2.to_csv("Datasets/all_nba_seasons_revised.csv")
```

```python
# Read the CSV file for players who won MVP and were in the MVP race for each season into a Pandas Dataframe

mvp_candidates = pd.read_csv("Datasets/mvps.csv", encoding='unicode_escape')
mvp_candidates
```

```python
# Filter out dataframe for our desired seasons, starting from 1996-97 NBA season

mvp_candidates = mvp_candidates[mvp_candidates["Season"] >= "1996-97"]
mvp_candidates
```

```python
# Drop columns that already exist in the statistics dataframe. We are dropping these here because we will merge
# the two dataframes and don't want to duplicate columns

mvp_candidates = mvp_candidates.drop(columns = ["Age", "Tm", "G", "MP", "PTS", "TRB", "AST", "STL", "BLK", "FG%", "3P%", "FT%", "WS", "WS/48"])
mvp_candidates
```

```python
# Merge the two dataframes into one combined dataframe

nba_combined = pd.merge(mvp_candidates, nba_df2, on = ["Season", "Player"])
```

```python
nba_combined
```

```python
# Set the column order

column_order = ["Season", "Rank", "Player", "Pos", "Age", "G", "MP", "PTS", "AST", "TRB", "STL", "BLK", "FG%", "eFG%", "FT%", "PF", "WS", "WS/48","First", "Pts Won", "Pts Max", "Share", "MVP"]
```

```python
# Restructure the dataframe to the desired column order

nba_combined = nba_combined[column_order]
```

```python
# Display dataframe after adjusments are made

nba_combined
```

```python
# Using the "Rank" column, if the player has a Rank of 1, add a 1 to the MVP column, otherwise make MVP column 0

nba_combined["MVP"] = nba_combined["Rank"].apply(lambda x: '1' if x == '1' else '0')
```

```python
# Display combined dataframe after adjustments are made

nba_combined
```

```python
# Save the NBA combined DataFrame
nba_combined.to_csv("Datasets/mvps_stats_all_seasons_revised.csv")
```

```python
# Load the RAPTOR data
latest_raptor_df = pd.read_csv('Datasets/latest_RAPTOR_by_team.csv', usecols=['player_name', 'season', 'raptor_offense', 'raptor_defense', 'predator_offense', 'predator_defense'])
historical_raptor_df = pd.read_csv('Datasets/historical_RAPTOR_by_team.csv', usecols=['player_name', 'season', 'raptor_offense', 'raptor_defense', 'predator_offense', 'predator_defense'])

# Convert the 'season' column in RAPTOR dataframes to string
latest_raptor_df['season'] = latest_raptor_df['season'].astype(str)
historical_raptor_df['season'] = historical_raptor_df['season'].astype(str)

# Merge the RAPTOR dataframes
raptor_df = pd.concat([latest_raptor_df, historical_raptor_df]).drop_duplicates(subset=['player_name', 'season'])

# Adjust the 'Season' column in nba_combined to match the 'season' format in RAPTOR data
def convert_season(season):
    year = season.split('-')[1]
    if int(year) < 50:  # Year cut off is 2050
        return '20' + year
    else:
        return '19' + year

nba_combined['Season'] = nba_combined['Season'].apply(convert_season)

# Merge the combined RAPTOR data with the nba_combined DataFrame
nba_combined = nba_combined.merge(raptor_df, left_on=['Player', 'Season'], right_on=['player_name', 'season'], how='left')

# Drop redundant columns
nba_combined.drop(['player_name', 'season'], axis=1, inplace=True)
```

```python
# Save the merged DataFrame
nba_combined.to_csv("Datasets/RAPTOR_all_seasons.csv")
```
