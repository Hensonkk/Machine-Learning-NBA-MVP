```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import os
import shutil
from splinter import Browser
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
```

## Scrapping 2022-2023 MVP's

```python
year = list(range(2022,2023))
```

```python
year
```

```python
mvp_url = "https://www.basketball-reference.com/awards/awards_{}.html"
```

```python
for x in year:
    url = mvp_url.format(x)
    
    data = requests.get(url)
    
    with open("mvp2022/{}.html".format(year), "w+") as f:
        f.write(data.text)
```

```python
with open("mvp2022/[2022].html") as f:
    page = f.read()
    
soup = BeautifulSoup(page, 'html.parser')
soup.find('tr', class_="over_header").decompose()
```

```python
mvp_table = soup.find_all(id="div_mvp")[0]
```

```python
mvps_2022 = pd.read_html(str(mvp_table))[0]
```

```python
mvps_2022.head()
```

```python
mvps_2022["Year"] = 2022
```

```python
mvps_2022.head()
```

```python
mvps_2022.to_csv("mvps2022_2023.csv")
```

## Scrapping 2022-2023 Player Stats

```python
# creates path to chrome 
browser = Browser('chrome')
```

```python
stats_2022_url = "https://www.basketball-reference.com/leagues/NBA_2023_per_game.html"
```

```python
# activates path to chrome (python has control of page)
browser.visit(stats_2022_url)
```

```python
year = list(range(2023,2024))
```

```python
year
```

```python
for x in year:
    url = stats_2022_url.format(year)
    
    data = requests.get(stats_2022_url)
    
    with open("stats_2022/{}.html".format(year), "w+") as f:
        f.write(data.text)
```

```python
with open("stats_2022/[2023].html") as f:
    page = f.read()
    
soup = BeautifulSoup(page, 'html.parser')
```

```python
for tr in soup.find_all('tr', class_="sortable stats_table now_sortable sticky_table eq2 re2 le2"):  # Replace 'your_class_name' with the actual class if needed
    tr.decompose()
```

```python
player_table = soup.find_all(id="div_per_game_stats")[0]
```

```python
player_table
```

```python
players_2022 = pd.read_html(str(player_table))[0]
```

```python
players_2022["Year"] = 2022
```

```python
players_2022.head()
```

```python
players_2022.to_csv("stats_2022_2023.csv")
```
