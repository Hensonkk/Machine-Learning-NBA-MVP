```python
year = list(range(2022,2023))
```

```python
year
```

```python
url_start = "https://www.basketball-reference.com/awards/awards_{}.html"
```

```python
for x in year:
    url = url_start.format(x)
    
    data = requests.get(url)
    
    with open("mvp2022/{}.html".format(year), "w+") as f:
        f.write(data.text)
```

```python
with open("mvp2022/[2022].html") as f:
    page = f.read()
    
soup = BeautifulSoup(page, 'html.parser')
soup.find('tr', class_="over_header").decompose()
```

```python
mvp_table = soup.find_all(id="div_mvp")[0]
```

```python
mvps_2022 = pd.read_html(str(mvp_table))[0]
```
